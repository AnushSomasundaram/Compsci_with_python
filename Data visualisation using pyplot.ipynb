```python
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(0.,10,0.1)
a=np.cos(x)
b=np.sin(x)
plt.plot(x,a,'b')
plt.plot(x,b,'r')
plt.show()
```

```python
#to change line width you can give an additional argument in plot(...) -> linewidth=<line width>
#to change line style " " " " " " " "                   " "             -> linestyle="<line style >"
# line styles -> solid , dashed , dasheddot , dotted
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(0.,10,0.1)
a=np.cos(x)
b=np.sin(x)
plt.plot(x,a,'b',linewidth = 4,linestyle="dashdot")
plt.plot(x,b,'r',linewidth=4,linestyle="dashed")
plt.show()

```

```python
#data points being plotted are called markers...to change the type of marker size,type,color we can use commands_>
#marker=<valid marker type>, markersize=<in points>, markeredgecolor=<valid color>
```

```python
plt.plot(a,b,marker="d",markersize=9,markeredgecolor="r")

```

```python
#Example 8.1 from data visualiztion using python (computer science with python)

import matplotlib.pyplot as plt
import numpy as np


x=np.arange(1,20,1.25)
y=np.log(x)
plt.plot(x,y)

plt.xlabel('Random Values')
plt.ylabel("Logarithm Values")

z=np.cos(x)
plt.plot(x,z,'b',linestyle='dashdot',marker='.',markeredgecolor='blue')


```
