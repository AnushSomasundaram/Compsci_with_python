```python
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(0.,10,0.1)
a=np.cos(x)
b=np.sin(x)
plt.plot(x,a,'b')
plt.plot(x,b,'r')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(0.,10,0.1)
a=np.cos(x)
b=np.sin(x)
plt.plot(x,a,'b',linewidth = 4,linestyle="dashdot")
plt.plot(x,b,'r',linewidth=4,linestyle="dashed",color="purple")
plt.show()

```

```python
#creating mutiple data sets on a bar chart for comparison
#we can create multiple bar charts based on 


import matplotlib.pyplot as plt
import numpy as np

A=[2,4,6,8]
B=[2.8,3.5,6.5,7.7]

"""
First we have to decide the number of x points...ie. we have to create a list of position elements for the values 
in the data set A.
We can use the range function or the numpy.arange() function to do so.....we acquire a set of x values which we 
can use to position the bars....hence we can say that the first data set lies at the value of x on the x axis and 
the  second data set lies on the x axis at a value of x(x+width) plus the width of the first bar
we than specify different color" " parameters for bars from the two data sets to be able to
distinguish between them

"""


x=np.arange(len(A))
print(x)
plt.bar(x,A,color="red",width=0.35)
plt.bar(x+0.35,B,color="blue",width=0.35)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.axis("equal")
contri=[17,8.8,12.75,14]
plt.pie(contri)

plt.show()
```

```python
#we can add labels to different slices of the pie chart
import matplotlib.pyplot as plt

plt.axis("equal")


contri=[17,8.8,12.75,14]

houses=["Vidya","Kshama","Namrta","Karuna"]

plt.pie(contri,labels=houses)
plt.show()



```

```python
# we can add slice percentages to the pie
#To view percentage shares of the pie , we need to add an argument called autopct with a format string, such as 
# "%1.1F%%"

#autopct="%1.1F%%"

plt.axis("equal")


contri=[17,8.8,12.75,14]

houses=["Vidya","Kshama","Namrta","Karuna"]

plt.pie(contri,labels=houses,autopct="%1.1F%%")

plt.show()


#theres an entire two pages of format strings in the book im just gonna skip it
```

```python
#to change the colours of each slice to prefered colours in the pie chart we can use the colors argument


plt.axis("equal")


contri=[17,8.8,12.75,14]

houses=["Vidya","Kshama","Namrta","Karuna"]

colrs=["Blue","Red","Green","Yellow"]

plt.pie(contri,labels=houses,colors=colrs,autopct="%1.1F%%")
 
plt.show()


```

```python
#we can explode a slice...ie move the slice away from the pie by using the explode argument
#first we have to create a list of how many units you want the slice to move away from the pie
#then you have to specify the argument explode=(name of list with units) in the brackets



plt.axis("equal")


contri=[17,8.8,12.75,14]

houses=["Vidya","Kshama","Namrta","Karuna"]

colrs=["Blue","Red","Green","Yellow"]

expl=[0.0,0.1,0.0,0.0]

plt.pie(contri,labels=houses,colors=colrs,explode=expl,autopct="%1.1F%%")

plt.show()

#as you can see the red slice has moved away from the pie

```

```python
#when data values are given pyplot automatically finds the suitable ranges but sometimes 
#we have to specify what the range of the x and y axes have to be
#hence we can use the xlim() and ylim() functions from the pyplot module

import matplotlib.pyplot as plt
import numpy as np

x=np.arange(4)
y=[1,5,6,7]

plt.xlim=[-2.0,4.0]

plt.bar(x,y)
plt.title("A bar chart")

# we can set up ticks using plt.xticks,plt.yticks and we can name the ticks by giving a list of names as a argument

plt.xticks([0,1,2,3],["A","B","C","D"])
plt.show()
```

```python
#we can also set up a legend by labeling all of the bars using the label="" argument
#and the plt.legend function inside the brackets you can also specify the location of the legend


val=[[5.,25.,45.,20.],[4.,23.,49.,17.],[6.,22.,47.,19.]]

x=np.arange(len(val[0]))

figure=plt.figure(figsize=(10,10))
plt.bar(x,val[0],width=0.25,color="red",label='range3')
plt.bar(x+0.25,val[1],width=0.25,color="purple",label='range1')
plt.bar(x+0.50,val[2],width=0.25,color="green",label='range2')
plt.xlabel("pointless values",color="red")
plt.ylabel("why",size=100,color="cyan")
plt.title("This graph has no point , but its the solution for example 8.2",)
plt.legend(loc="upper left")

plt.show()
```
